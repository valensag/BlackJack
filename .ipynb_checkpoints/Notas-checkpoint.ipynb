```python
'''
Game introduction
Welcome message, game instructions, ask the user if they want to start the game.
'''
from IPython.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams
from random import randint
from IPython.display import Markdown, display

class Introduction_Game:
    def __init__(self,card1_dealer = '', card2_dealer = '', amount = 0, turn = 0):
        self.card1_dealer = card1_dealer
        self.card2_dealer = card2_dealer
        self.amount = amount
        self.turn = turn
        display(Markdown('**Welcome to BLACKJACK!**'))
        display(Markdown('In this game, the winner will be the one whose sum of the value of the cards is closest to **21**. Whoever exceeds this amount automatically loses.'))
    
    def instructions(self):
        display(Markdown('**This are the Instructions:**'))
        print('1. Determine the bet amount (5,10,25,100). \n2. Two cards will be given to you. You must choose between the following actions to perform:')
        print('    a. HIT -----> Hit another card.\n    b. PLANT ---> End the turn and play with the current hand.\n    c. DOUBLE --> Double your bet.')
        print("3. Dealer's turn.")
        print('4. Players who stay further than 21 than the dealer or who have exceeded this value lose.')
        print('5. Those who have the same dealer value recover their bet but do not lose or win.')
    
    def start_game(self):
        '''
        Input: User question
        Output: Answer about to start or not the game
        '''
        while True:
            try:
                start = int(input('Start? [Yes:1 / No:0]  '))
                return(start)
                break
            except:
                print('Sorry, You have to enter 1 or 0')
                continue
                
    def show_bet_image(self):
        '''
        Show bet image
        '''
        %matplotlib inline
        rcParams['figure.figsize'] = 8,2.5
        bet_image = mpimg.imread('/Users/usuario/Desktop/blackjack_images/bet.png')
        plt.suptitle('1                          2                            3                            4')
        plt.imshow(bet_image)
        plt.show()
            
    def bet_amount(self):
        while True:
                try:
                    answer = int(input('How much are you going to bet? '))
                    break
                except:
                    print('Sorry, You have to enter a number between 1 and 4')
                    continue
        if answer == 1:
            self.amount = 5
        elif answer == 2:
            self.amount = 10
        elif answer == 3:
            self.amount = 25
        else:
            self.amount = 100
        return(self.amount)
    
    def dealer_aditional_move(self):
        
        display(Markdown('**DEALER GAME**'))
        
        %matplotlib inline
        rcParams['figure.figsize'] = 5,2.5

        self.card1_dealer = str(randint(0,13)) + "-" + str(randint(0,3))
        self.card2_dealer = 'Carta_reverso'
        
        picture1 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.card1_dealer}.png')
        picture2 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.card2_dealer}.png')

        fig, ax = plt.subplots(1,2)
        ax[0].imshow(picture1)
        ax[1].imshow(picture2)
        plt.show()
        
        return(self.card1_dealer,self.card2_dealer)
    
    def dealer_turn(self):
        
        %matplotlib inline
        rcParams['figure.figsize'] = 5,2.5

        self.card2_dealer = str(randint(0,13)) + "-" + str(randint(0,3))
        
        picture1 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.card1_dealer}.png')
        picture2 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.card2_dealer}.png')

        fig, ax = plt.subplots(1,2)
        ax[0].imshow(picture1)
        ax[1].imshow(picture2)
        
        self.card1_dealer = self.card1_dealer.split('-')[0]
        self.card2_dealer = self.card2_dealer.split('-')[0]
        
        plt.show()
        
        return(self.card1_dealer,self.card2_dealer)
    
        
    def player_turn(self):
        
        display(Markdown('**PLAYER GAME**'))
        
        %matplotlib inline
        rcParams['figure.figsize'] = 5,2.5

        card1 = str(randint(0,13)) + "-" + str(randint(0,3))
        card2 = str(randint(0,13)) + "-" + str(randint(0,3))
        
        picture1 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{card1}.png')
        picture2 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{card2}.png')

        fig, ax = plt.subplots(1,2)
        ax[0].imshow(picture1)
        ax[1].imshow(picture2)
        
        card1 = card1.split('-')[0]
        card2 = card2.split('-')[0]
        
        print('card1:',card1)
        print('card2:',card2)
        
        plt.show()
        
        return(card1,card2)
        
    
    def hit_plant_double(self):
        print('Choose from the following options: ')
        while True:
                try:
                    response = int(input('1.HIT \n2.PLANT \n3.DOUBLE \n'))
                    break
                except:
                    print('Sorry, You have to enter a number between 1 and 3')
                    continue
        return(response)
        
    def hit(self):
        %matplotlib inline
        rcParams['figure.figsize'] = 5,2.5
        
        aditional_card = str(randint(0,13)) + "-" + str(randint(0,3))
        picture1 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{aditional_card}.png')
        plt.imshow(picture1)
        
        aditional_card = aditional_card.split('-')[0]
        return(aditional_card)
    
    def plant(self):
        self.turn = 1
        return(self.turn)
    
    def double(self):
        self.amount *= 2
        return(self.amount)
        
        
                
```

```python
'''
Game development
Inherit the class Introduction_game and it's initialization
'''
from IPython.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams
from random import randint

class Develope_Game(Introduction_Game):
    
    def __init__(self,total_amount = 1000,player_board = [],dealer_board = [],count_player = 0,count_dealer = 0,turn = 0):
        self.total_amount = total_amount
        self.player_board = player_board
        self.count_player = count_player
        self.count_dealer = count_dealer
        self.turn = turn
        
        Introduction_Game.__init__(self)
        Introduction_Game.instructions(self)
        
        start = Introduction_Game.start_game(self)
        
        if start == 1:
            bet = Introduction_Game.show_bet_image(self)
        else:
            print('Good bye!')
        
    def bet_amount_develope(self):
        amount = Introduction_Game.bet_amount(self)
        print('\nTotal amount: ',self.total_amount)
        print('Actual bet: ', amount)
        Introduction_Game.dealer_aditional_move(self)

    #####
    def cards_of_player(self):
        card1,card2 = Introduction_Game.player_turn(self)
        self.player_board.append(int(card1))
        self.player_board.append(int(card2))
        return(self.player_board)
    
    def count_cards_of_player(self):
        self.count_player = 0
        for item in range(len(self.player_board)):
            self.count_player += self.player_board[item]
        return(self.count_player)
    ####
    def cards_of_dealer(self):
        card1,card2 = Introduction_Game.dealer_turn(self)
        self.dealer_board.append(int(card1))
        self.dealer_board.append(int(card2))
        return(self.dealer_board)
    
    def count_cards_of_dealer(self):
        self.count_dealer = 0
        for item in range(len(self.dealer_board)):
            self.count_dealer += self.dealer_board[item]
        return(self.count_dealer)
    ####
    
    def comparation_to_21(self):
        if self.count_player > 21:
            print('OVERFLOW! You exceeded the number of 21')
            play_again = input('Play Again? [Y:1 / N:0]')
            return(play_again)
        elif self.count_player == 21:
            print('WINNER!!')
            return('winner')
        else:
            print('Hora de decidir!')
            response = Introduction_Game.hit_plant_double()
            return(response)
    
    def Add_aditional_card(self):
        #Este método trabaja con HIT
        aditional_card = Introduction_Game.hit(self)
        if self.turn == 0:
            self.player_board.append(int(aditional_card))
        else:
            self.dealer_board.append(int(aditional_card))
            
        print(self.player_board)
            
        
        
        
```

```python
'''
Unión de todo el código
'''
class Play_Game(Develope_Game):
    
    def __init__(self):
        Develope_Game.__init__(self)
        Develope_Game.bet_amount_develope(self)
        Develope_Game.cards_of_player(self)
    
```
