```python
import import_ipynb
import module_dealer
import module_player
dealer = module_dealer.Class_Dealer()
player = module_player.Class_Player()

from IPython.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams
from random import randint
from IPython.display import Markdown, display
from IPython.display import clear_output
import imageio
import time

class General():
    def __init__(self,start = '', turn = 0, winner = ''):
        self.winner = winner
        self.start = start
        self.turn = turn
        display(Markdown('**Welcome to BLACKJACK!**'))
        display(Markdown('In this game, the winner will be the one whose sum of the value of the cards is closest to **21**. Whoever exceeds this amount automatically loses.'))
        
    def instructions(self):
        display(Markdown('**INSTRUCTIONS:**'))
        print('1. Determine the bet amount (5,10,25,100). \n2. Two cards will be given to you. You must choose between the following actions to perform:')
        print('    a. HIT -----> Hit another card.\n    b. PLANT ---> End the turn and play with the current hand.\n    c. DOUBLE --> Double your bet.')
        print("3. Dealer's turn.")
        print('4. Players who stay further than 21 than the dealer or who have exceeded this value lose.')
        print('5. Those who have the same dealer value recover their bet but do not lose or win.')
    
    def start_game(self):
        while True:
            try:
                self.start = int(input('Start? [Yes:1 / No:0]  '))
                return(self.start)
                break
            except:
                print('Sorry, You have to enter 1 or 0')
                continue
        
    def show_bet_image(self):
        clear_output()
        display(Markdown('**BET AMOUNT**'))
        %matplotlib inline
        rcParams['figure.figsize'] = 8,2.5
        bet_image = mpimg.imread('/Users/usuario/Desktop/blackjack_images/bet.png')
        plt.suptitle('1                          2                            3                            4')
        plt.imshow(bet_image)
        plt.show()
        player.player_bet()
        
    def start_player_turn(self, cont = True, continuar = True):
        self.cont = cont
        self.continuar = continuar
        dealer.dealer_first_turn()
        player.player_first_turn()
        player.player_score()
        
        while self.cont:
            player.hit_plant_double()
            
            if player.response == 1:
                player.aditional_card()
                player.mazo()
                player.player_score()
                if player.score > 21:
                    self.winner = 'Dealer'
                    self.cont = False
                    self.continuar = False
                elif player.score == 21:
                    self.winner = 'Player'
                    self.cont = False
                    self.continuar = False
                else:
                    continue
                
            elif player.response == 2:
                self.cont = False
                self.turn = 1
                
            elif player.response == 3:
                player.bet *= 2
                print('Total amount: ', player.total_amount)
                print('Your bet: ', player.bet)
                self.cont = False
                self.turn = 1
                
    def start_dealer_turn(self, contador = True, win = 0 ):
        self.win = win
        self.contador = contador
        if self.continuar == True:
            if self.turn == 1:
                dealer.dealer_bet()
                dealer.dealer_first_turn()
                dealer.dealer_next_turn()
                dealer.dealer_score() 
            else:
                pass

            while self.contador:
                dealer.hit_plant_double()
                time.sleep(3)
                if dealer.respuesta == 1:
                    dealer.aditional_card()
                    dealer.mazo()
                    dealer.dealer_score()
                    if dealer.puntaje > 21:
                        self.winner = 'Player'
                        self.contador = False
                    elif dealer.puntaje == 21:
                        self.winner = 'Dealer'
                        self.contador = False
                    else:
                        continue
                elif dealer.respuesta == 2:
                    self.win = 1
                    self.contador = False
                    self.continuar = False

                elif dealer.respuesta == 3:
                    dealer.apuesta = player.bet
                    
                    self.win = 1
                    self.contador = False
                    self.continuar = False

        else:
            pass
            
    def select_winner(self):
        if self.win == 1:
            print('Dealer score: ', dealer.puntaje)
            print('Player score: ', player.score)
            resta = player.score - dealer.puntaje
            if resta > 0:
                display(Markdown('Your score is closer to **21** \n**YOU WIN!** Congratulations! \U0001F60A'))
                self.winner = 'Player'
            elif resta < 0 :
                display(Markdown('Your score is farthest from **21** \n**DEALER WIN!** A computer beat you! \U0001F923'))
                self.winner = 'Dealer'
            else:
                display(Markdown('**Tie! No one wins**'))
        else:
            pass
        
    def amount(self):
        if self.winner == 'Player':
            player.total_amount += player.bet
            print('Total amount Player: ', player.total_amount)
    
            dealer.total_cantidad -= player.bet
            print('Total amount Dealer: ', dealer.total_cantidad) 
        elif self.winner == 'Dealer':
            player.total_amount -= player.bet
            print('Total amount Player: ', player.total_amount)
            dealer.total_cantidad += player.bet
            print('Total amount Dealer: ', dealer.total_cantidad)
        else:
            pass
        return(player.total_amount, dealer.total_cantidad)
        
    def play_again(self, again = 0):
        self.again = again
        while True:
                try:
                    self.again = int(input('\nPlay again? [Y:1 / N:0]'))
                    break
                except:
                    print('Sorry, You have to enter 0 or 1')
                    continue
        return(self.again)
```
