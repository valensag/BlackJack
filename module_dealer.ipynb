```python
from IPython.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams
from random import randint
from IPython.display import Markdown, display
from IPython.display import clear_output
import imageio

class Class_Dealer():
    
    def __init__(self,carta1 = '', carta2 = '',respuesta = 0, numero_cartas = 2, espacio = 5,tablero = [],imagen1 = '', imagen2 = '', cuenta_lista = [],apuesta = 0,
                 total_cantidad = 1000):
        self.carta1 = carta1
        self.carta2 = carta2
        self.respuesta= respuesta
        self.numero_cartas = numero_cartas
        self.espacio = espacio
        self.tablero = tablero
        self.imagen1 = imagen1
        self.imagen2 = imagen2
        self.cuenta_lista = cuenta_lista 
        self.apuesta = apuesta
        self.total_cantidad = total_cantidad
        
        
    def dealer_bet(self):
        print('\nTotal amount Dealer: ', self.total_cantidad)
        
    def dealer_first_turn(self):
        display(Markdown('**DEALER TURN**'))
        
        %matplotlib inline
        rcParams['figure.figsize'] = 5,2.5
        
        self.carta1 = str(randint(1,13)) + "-" + str(randint(1,3))
        self.carta2 = 'Carta_reverso'
        

        imagen1 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.carta1}.png')
        imagen2 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.carta2}.png')
        
        plt.imsave('imagen1.jpg',imagen1)
        self.imagen1 = imageio.imread('imagen1.jpg')
        
        fig, ax = plt.subplots(1,self.numero_cartas)
        ax[0].imshow(imagen1)
        ax[1].imshow(imagen2)
        plt.show()
    
        
    def dealer_next_turn(self):
        clear_output()
        display(Markdown('**DEALER TURN**'))
        %matplotlib inline
        rcParams['figure.figsize'] = 5,2.5
        self.carta2 = str(randint(1,13)) + "-" + str(randint(1,3))
        imagen2 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.carta2}.png')
        
        fig, ax = plt.subplots(1,self.numero_cartas)
        ax[0].imshow(self.imagen1)
        ax[1].imshow(imagen2)
        plt.show()
        
        plt.imsave('imagen2.jpg',imagen2)
        self.imagen2 = imageio.imread('imagen2.jpg')
        self.tablero.extend([self.imagen1,self.imagen2])
        
        self.carta1 = self.carta1.split('-')[0]
        self.carta2 = self.carta2.split('-')[0]
        
        self.cuenta_lista.extend([self.carta1,self.carta2])
        return(self.carta1, self.carta2)
    
    def dealer_score(self, puntaje = 0):
        self.puntaje = puntaje
        #print(self.cuenta_lista)
        for i in range(len(self.cuenta_lista)):
            number = int(self.cuenta_lista[i])
            if number > 10:
                subtraction = number - 10
                number -= subtraction
                self.cuenta_lista[i] == number
            else:
                pass
                
            self.puntaje+= number
            if self.puntaje > 21:
                print('DEALER LOSE! exceeded 21!')
                display(Markdown('**YOU WIN!** Congratulations! \U0001F60A')) 
                break
            elif self.puntaje == 21:
                display(Markdown('**DEALER WIN!** A computer beat you! \U0001F923'))
                break
            else:
                continue
                
        print('SCORE: ',self.puntaje)
        
    def hit_plant_double(self):
#         print('\nChoose from the following options: ')
#         while True:
#                 try:
#                     self.respuesta = int(input('1.HIT \n2.PLANT \n3.DOUBLE \n'))
#                     break
#                 except:
#                     print('Sorry, You have to enter a number between 1 and 3')
#                     continue
        if self.puntaje > 19:
            self.respuesta = 3
        elif self.puntaje >= 18 and self.puntaje <20:
            self.respuesta = 2
        else:
            self.respuesta = 1
    
    def aditional_card(self,imagen_nueva = ''):
        if self.respuesta == 1:
            clear_output()
          
        display(Markdown('**DEALER TURN**'))
        
        self.imagen_nueva = imagen_nueva
        self.numero_cartas += 1
        
        %matplotlib inline
        rcParams['figure.figsize'] = self.espacio,2.5
        
        nueva_carta = str(randint(1,13)) + "-" + str(randint(1,3))
        imagen_nueva = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{nueva_carta}.png')
        
        plt.imsave('imagen_nueva.jpg',imagen_nueva)
        self.imagen_nueva = imageio.imread('imagen_nueva.jpg')
        
        plt.imshow(self.imagen_nueva)
        self.tablero.extend([self.imagen_nueva])
        
        nueva_carta = nueva_carta.split('-')[0]
        self.cuenta_lista.append(nueva_carta)
    
    def mazo(self):
        
        self.espacio += 2
        if self.numero_cartas > 4:
            self.espacio += 2
        %matplotlib inline
        rcParams['figure.figsize'] = self.espacio,2.5
        
        fig,ax = plt.subplots(1,self.numero_cartas)
        for item in range(self.numero_cartas):
            ax[item].imshow(self.tablero[item])
        plt.show()
        
            
```
