```python
from IPython.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams
from random import randint
from IPython.display import Markdown, display
from IPython.display import clear_output
import imageio


class Class_Player():
    
    def __init__(self,card1 = '', card2 = '',response = 0, num_cards = 2, espaciado = 5,board = []):
        self.card1 = card1
        self.card2 = card2
        self.response = response
        self.num_cards = num_cards
        self.espaciado = espaciado
        self.board = board
        
    def player_bet(self,bet = 0, total_amount = 1000):
        self.bet = bet
        self.total_amount = total_amount
        while True:
            try:
                answer = int(input('\nEnter the number that corresponds to your selected bet:  '))
                if answer <= 4:
                    break
                else:
                    print('Error! Enter a number between 1 and 3')
            except:
                print('Error! Enter a number between 1 and 3')
                continue
        
        clear_output()
        if answer == 1:
            self.bet = 5
        elif answer == 2:
            self.bet = 10
        elif answer == 3:
            self.bet = 25
        else:
            self.bet = 100
        
        print('\nTotal amount Player: ', self.total_amount)
        print('Your bet: ', self.bet)
        
        
    def player_first_turn(self, picture1 = '', picture2 = '',count_list = []):
        self.picture1 = picture1
        self.picture2 = picture2
        self.count_list = count_list
        
        display(Markdown('**PLAYER TURN**'))
        
        %matplotlib inline
        rcParams['figure.figsize'] = 5,2.5
        
        self.card1 = str(randint(1,13)) + "-" + str(randint(1,3))
        self.card2 = str(randint(1,13)) + "-" + str(randint(1,3))

        picture1 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.card1}.png')
        picture2 = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{self.card2}.png')
        
        fig, ax = plt.subplots(1,self.num_cards)
        ax[0].imshow(picture1)
        ax[1].imshow(picture2)
        plt.show()
        
        plt.imsave('picture1.jpg',picture1)
        self.picture1 = imageio.imread('picture1.jpg')
        plt.imsave('picture2.jpg',picture2)
        self.picture2 = imageio.imread('picture2.jpg')
        self.board.extend([self.picture1,self.picture2])
        
        self.card1 = self.card1.split('-')[0]
        self.card2 = self.card2.split('-')[0]
        
        self.count_list.extend([self.card1,self.card2])
        return(self.card1, self.card2)
    
    def player_score(self, score = 0):
        self.score = score
        #print(self.count_list)
        for i in range(len(self.count_list)):
            number = int(self.count_list[i])
            if number > 10:
                subtraction = number - 10
                number -= subtraction
                self.count_list[i] == number
            else:
                pass
                
            self.score += number
            if self.score > 21:
                print('YOU LOSE! you exceeded 21!')
                display(Markdown('**DEALER WIN!** A computer beat you! \U0001F923'))
                break
            elif self.score == 21:
                
                display(Markdown('**YOU WIN!** Congratulations! \U0001F60A'))
                break
            else:
                continue
                
        print('SCORE: ',self.score)

    
    def hit_plant_double(self):
        print('\nChoose from the following options: ')
        while True:
                try:
                    self.response = int(input('1.HIT \n2.PLANT \n3.DOUBLE \n'))
                    break
                except:
                    print('Sorry, You have to enter a number between 1 and 3')
                    continue
        #return(self.response)
            
    def aditional_card(self,picture_new = ''):
        if self.response == 1:
            clear_output()
          
        display(Markdown('**PLAYER TURN**'))
        
        self.picture_new = picture_new
        self.num_cards += 1
        
        %matplotlib inline
        rcParams['figure.figsize'] = self.espaciado,2.5
        
        new_card = str(randint(1,13)) + "-" + str(randint(1,3))
        picture_new = mpimg.imread(f'/Users/usuario/Desktop/blackjack_images/{new_card}.png')
        
        plt.imsave('picture_new.jpg',picture_new)
        self.picture_new = imageio.imread('picture_new.jpg')
        
        plt.imshow(self.picture_new)
        self.board.extend([self.picture_new])
        
        new_card = new_card.split('-')[0]
        self.count_list.append(new_card)
        
    def mazo(self):
        
        self.espaciado += 2
        if self.num_cards > 4:
            self.espaciado += 2
        %matplotlib inline
        rcParams['figure.figsize'] = self.espaciado,2.5
        
        fig,ax = plt.subplots(1,self.num_cards)
        for item in range(self.num_cards):
            ax[item].imshow(self.board[item])
        plt.show()
            
```
